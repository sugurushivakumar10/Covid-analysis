```python
plt.figure(figsize=(18,18))
sns.barplot(x="state_name",y="active",data=covid)
plt.show()
```

## Active cases in India

```python
plt.figure(figsize=(12,8))
ax = sns.barplot(x="state_name", y="active", data=covid, palette=sns.color_palette("GnBu", 10))
plt.xticks(rotation='vertical')
ax.grid(b=True, which='major', color='#d3d3d3', linewidth=1.0)
ax.grid(b=True, which='minor', color='#d3d3d3', linewidth=0.5)
plt.show()
```

## Positive cases in India

```python
plt.figure(figsize=(12,8))
ax = sns.barplot(x="state_name", y="positive", data=covid, palette=sns.color_palette("GnBu", 10))
plt.xticks(rotation='vertical')
ax.grid(b=True, which='major', color='#d3d3d3', linewidth=1.0)
ax.grid(b=True, which='minor', color='#d3d3d3', linewidth=0.5)
plt.show()
```
